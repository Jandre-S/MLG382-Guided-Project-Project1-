```python
import warnings 
warnings.simplefilter("ignore",category=FutureWarning)
import pandas as pd
from sklearn.impute import SimpleImputer
raw_df = pd.read_csv('../data/raw_data.csv')
raw_df.info()
```

```python
raw_df.isnull().sum()
```

```python
val = ['Gender', 'Married','Education','Self_Employed','Preperty_Area','Loan_Status']
processed_df = raw_df
for x in range(0,4):
    
    recode_cat_values_df = pd.get_dummies(
        processed_df, 
        columns=[val[x]], 
        dtype=int) 
    processed_df = recode_cat_values_df
recode_cat_values_df.head(200)
```

```python
missing = ['LoanAmount', 'Loan_Amount_Term','Credit_History','Dependents']

for y in range(0,3):
    processed_df.loc[
        processed_df[missing[y]].isnull()==True,# row location where value == True for NaN
        missing[y] # col location
        ] = processed_df[missing[y]].mean()

processed_df.head(200)

# Checking missing values again
```

```python

irrelevant_features = ['Married_No', 'Education_Not Graduate', 'Self_Employed_No']

# Remove irrelevant features
new_df2 = processed_df
new_df2.drop(
    columns=irrelevant_features,
    inplace=True
)
new_df2.head()
```
