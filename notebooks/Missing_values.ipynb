```python
import warnings 
warnings.simplefilter("ignore",category=FutureWarning)
import pandas as pd
from sklearn.impute import SimpleImputer
raw_df = pd.read_csv('../data/raw_data.csv')
raw_df.head()
```

```python
raw_df.isnull().sum()
```

```python
val = ['Gender', 'Married','Education','Self_Employed','Preperty_Area','Loan_Status']
processed_df = raw_df
for x in range(0,4):
    
    recode_cat_values_df = pd.get_dummies(
        processed_df, 
        columns=[val[x]], 
        dtype=int) 
    processed_df = recode_cat_values_df
recode_cat_values_df.head()
```

```python

processed_df.loc[
    processed_df['LoanAmount'].isnull()==True,# row location where value == True for NaN
    'LoanAmount' # col location
    ] = processed_df['LoanAmount'].mean()

processed_df.head()

# Checking missing values again
```

```python

irrelevant_features = ['Married_No', 'Education_Not Graduate', 'Self_Employed_No']

# Remove irrelevant features
new_df2 = processed_df
new_df2.drop(
    columns=irrelevant_features,
    inplace=True
)
new_df2.head()
```
